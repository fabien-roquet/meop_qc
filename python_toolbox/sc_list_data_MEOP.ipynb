# create a list of all profiles

```python
from os import walk
from os.path import isfile, join
import glob
import pandas as pd
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import gsw

processdir = '/home/jupyter-froqu/MEOP_process'
datadir = join(processdir,'final_dataset_prof')
```

```python
def N_PARAM(ds,PARAM):
    if PARAM+'_QC' in list(ds.variables):
        N_PARAM = np.sum((ds[PARAM+'_QC']==b'1'),axis=1)
    else:
        N_PARAM = xr.DataArray(np.zeros(ds.dims['N_PROF']), dims=['N_PROF'])
    return N_PARAM


list_df=[]
for (dirpath, dirnames, filenames) in walk(datadir):
    for dirname in dirnames:
        print(dirname)
        list_ncfile = glob.glob(join(dirpath,dirname,'{}-*_lr0_prof.nc'.format(dirname)))       
        for ncfile in list_ncfile:
            with xr.open_dataset(ncfile) as ds:
                data = {'SMRU_PLATFORM_CODE': ds.smru_platform_code,
                        'CYCLE_NUMBER': ds['CYCLE_NUMBER'].astype(int),
                        'JULD': ds['JULD'],
                        'LATITUDE': ds['LATITUDE'],
                        'LONGITUDE': ds['LONGITUDE'],
                        'N_TEMP' : N_PARAM(ds,'TEMP'),
                        'N_PSAL' : N_PARAM(ds,'PSAL'),
                        'N_CHLA' : N_PARAM(ds,'CHLA')}
            df = pd.DataFrame(data).set_index(['SMRU_PLATFORM_CODE','CYCLE_NUMBER'])
            list_df.append(df)
df_all = pd.concat(list_df)
```

```python
df_all.to_pickle('list_meop.pkl')
#df_all.read_pickle('list_meop.pkl')
```
